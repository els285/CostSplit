```python
shop1 = 125
def divide_based_on_weight(amount,dic_of_weights):
    
    sum_of_weights = sum(list(dic_of_weights.values()))
    
    new_dic = {}
    
    for person,w in dic_of_weights.items():
        new_dic[person] = w/sum_of_weights * amount
    
    return new_dic
```
